```python
import PySimpleGUI as sg

L1 = [[sg.Button('A', size=(5,1)), sg.Button('B', size=(5,1))],
      [sg.Button('C', size=(5,1)), sg.Button('D', size=(5,1))]]

L2 = [[sg.Button('E', size=(5,1)), sg.Button('F', size=(5,1))],
      [sg.Button('G', size=(5,1)), sg.Button('H', size=(5,1))]]

L = [[sg.Frame('Group 1', L1), sg.Frame('Group 2', L2)]]

window = sg.Window('sample01', L)

while True:
    event, values = window.read()
    print('イベント: ', event, '値: ', values)
    if event == None:
        break

window.close()
```

```python
import PySimpleGUI as sg

bts = ('1', '2', '3', '4', '5', '6', '7', '8', '9')

L = [[sg.Text('0000',  key='tx1')], 
     [sg.Button('-1', size=(5,3), key='bt1'), 
      sg.ProgressBar(100, orientation='h', size=(25, 25), key='pbar1', bar_color=('#aaa', '#fff'))]
    ]

window = sg.Window('sample03', L)
count = 0

while True:
    event, values = window.read(80)
    if event == None:
        break
    if event in (None, 'bt1'):
        count -= 3
    count += 1
    p1.UpdateBar(count);
    window['tx1'].Update(count);

window.close()
```

```python
import PySimpleGUI as sg
import random

random.seed(0)
bts = ('1', '2', '3', '4', '5', '6', '7', '8', '9')

# ボタンの固定キーを作成
keys = ['btn{}'.format(i) for i in range(9)]

# ボタンキーと数字のマッピングを初期化
key_to_num = {}

# ボタンをシャッフルして更新する関数
def shuffle_buttons(window):
    nums = list(range(1, 10))  # 1から9の数字
    random.shuffle(nums)
    for i, key in enumerate(keys):
        num = nums[i]
        key_to_num[key] = num
        window[key].update(text=str(num), disabled=False)

# レイアウトを作成（固定キーを使用）
L = [
    [sg.ProgressBar(100, orientation='h', size=(15, 25), key='pbar1', bar_color=('#aaa', '#fff')), sg.Text('', key='tx1')],
    [sg.Button('', size=(5, 3), key=keys[0]),
     sg.Button('', size=(5, 3), key=keys[1]),
     sg.Button('', size=(5, 3), key=keys[2])],
    [sg.Button('', size=(5, 3), key=keys[3]),
     sg.Button('', size=(5, 3), key=keys[4]),
     sg.Button('', size=(5, 3), key=keys[5])],
    [sg.Button('', size=(5, 3), key=keys[6]),
     sg.Button('', size=(5, 3), key=keys[7]),
     sg.Button('', size=(5, 3), key=keys[8])]
]

window = sg.Window('ボタン位置ランダムゲーム', L)
p1 = window['pbar1']
count = 100
i = 1  # 次に押すべき数字

# 最初にボタンをシャッフル
shuffle_buttons(window)

while True:
    event, value = window.read(timeout=50)

    if event is None:
        break

    if event != '__TIMEOUT__':
        if event in keys:
            num = key_to_num[event]
            if num == i:
                window[event].update(disabled=True)
                i += 1
                if i > 9:
                    if count > 0:
                        sg.popup(f'成功！残り時間は {count} です！')
                        break
                    else:
                        sg.popup('時間切れです。')
                        break
                else:
                    # 正しいボタンを押すたびにボタンをシャッフル
                    shuffle_buttons(window)
            else:
                sg.popup('間違ったボタンを押しました。')
                break

    count -= 1
    p1.UpdateBar(count)
    window['tx1'].Update(f'残り時間: {count}')

    if count <= 0:
        sg.popup('時間切れです。')
        break

window.close()
```
